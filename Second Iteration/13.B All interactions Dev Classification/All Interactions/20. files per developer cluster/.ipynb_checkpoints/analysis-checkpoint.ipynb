```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'correlation_levels_summary.csv'
data = pd.read_csv(file_path)

# Plotting function
def plot_correlation_levels(data):
    # Set the aesthetic style of the plots
    sns.set(style="whitegrid")

# Define the order of correlation levels
order = ["low", "medium", "strong"]

#Function to create heatmaps
def plot_heatmaps(data):
    # Get the unique classes
    classes = data['class'].unique()

    # Determine the number of rows and columns for subplots
    num_classes = len(classes)
    cols = 2
    rows = (num_classes + 1) // cols

    # Create a figure and a set of subplots
    fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(30, rows * 30), sharex=True, sharey=True)

    # Flatten the axes array for easy iteration
    axes = axes.flatten()

    # Loop through each class and create a heatmap
    for i, developer_class in enumerate(classes):
        # Filter data for the current class
        class_data = data[data['class'] == developer_class]

        # Pivot the data for the heatmap
        pivot_table = class_data.pivot_table(index='metric', columns='correlation_level', values='num_projects', fill_value=0)
        
        # Ensure the pivot table has all correlation levels
        for level in order:
            if level not in pivot_table.columns:
                pivot_table[level] = 0
        
        # Reorder the columns
        pivot_table = pivot_table[order]

        # Create a heatmap
        sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='viridis', ax=axes[i])

        # Set titles and labels
        axes[i].set_title(f"Correlation Levels for {developer_class}")
        axes[i].set_xlabel("Correlation Level")
        axes[i].set_ylabel("Metric")
        axes[i].tick_params(axis='y', rotation=0)
        axes[i].tick_params(axis='x', rotation=45)

        # Set y-ticks to be integers
        y_ticks = np.arange(len(pivot_table.index))
        axes[i].set_yticks(y_ticks)
        axes[i].set_yticklabels(pivot_table.index, rotation=0)

    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    # Adjust layout
    plt.tight_layout()


    # Show the plot
  
    plt.savefig("plot.png")
    plt.show()

# Plot the data
plot_heatmaps(data)
```

```python
import pandas as pd
import os
import re

# Function to determine the level of correlation
def correlation_level(pearson):
    if abs(pearson) >= 0.7:
        return 'strong'
    elif abs(pearson) >= 0.3:
        return 'medium'
    else:
        return 'low'

# Initialize an empty DataFrame to hold all the data
all_data = pd.DataFrame()

# Path to the directory containing the project files
directory_path = 'correlations'

# Regular expression pattern to extract project ID and developer class from the filename
pattern = re.compile(r'\d+_(\d+)_(.*?)\.csv')

# Loop through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith(".csv"):  # Ensure the file is a CSV
        file_path = os.path.join(directory_path, filename)
        project_data = pd.read_csv(file_path, names=['metric', 'pearson', 'sample_size'])
        
        # Extract project ID and developer class from the filename
        match = pattern.match(filename)
        if match:
            project_id = match.group(1)
            developer_class = match.group(2)
            
            # Add columns for project ID and developer class
            project_data['project'] = project_id
            project_data['class'] = developer_class
            
            # Append the data to the all_data DataFrame
            all_data = pd.concat([all_data, project_data], ignore_index=True)

# Transform the Pearson coefficient values into correlation levels
all_data['correlation_level'] = all_data['pearson'].apply(correlation_level)
all_data = all_data[all_data['metric'] != 'delta_action_delta_date']

# Count the number of times each correlation level appears for each metric and developer class
result = all_data.groupby(['class', 'metric', 'correlation_level']).size().reset_index(name='num_projects')

# Pivot the data to get the desired format
pivoted_data = result.pivot_table(index='metric', columns=['correlation_level', 'class'], values='num_projects', fill_value=0)
pivoted_data = pivoted_data.reindex(columns=pd.MultiIndex.from_product([['low', 'medium', 'strong'], ['super inactive', 'inactive', 'average', 'active', 'super active']]), fill_value=0)

# Split the data into separate tables for each correlation level
low_table = pivoted_data['low'].reset_index()
medium_table = pivoted_data['medium'].reset_index()
strong_table = pivoted_data['strong'].reset_index()

# Save the result to CSV files if needed
low_table.to_csv('correlation_levels_summary_low.csv', index=False)
medium_table.to_csv('correlation_levels_summary_medium.csv', index=False)
strong_table.to_csv('correlation_levels_summary_strong.csv', index=False)

# Save the result to LaTeX files if needed
low_table.to_latex('correlation_levels_summary_low.tex', index=False)
medium_table.to_latex('correlation_levels_summary_medium.tex', index=False)
strong_table.to_latex('correlation_levels_summary_strong.tex', index=False)

# Display the result
print("Low Correlation Table:")
print(low_table)
print("\nMedium Correlation Table:")
print(medium_table)
print("\nStrong Correlation Table:")
print(strong_table)
```

```python
import pandas as pd
import os
import re

# Function to determine the level of correlation
def correlation_level(pearson):
    if abs(pearson) >= 0.7:
        return 'strong'
    elif abs(pearson) >= 0.3:
        return 'medium'
    else:
        return 'low'

# Initialize an empty DataFrame to hold all the data
all_data = pd.DataFrame()

# Path to the directory containing the project files
directory_path = 'correlations'

# Regular expression pattern to extract project ID and developer class from the filename
pattern = re.compile(r'\d+_(\d+)_(.*?)\.csv')

# Dictionary to hold the unique project count for each class
project_class_count = {}

# Loop through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith(".csv"):  # Ensure the file is a CSV
        file_path = os.path.join(directory_path, filename)
        project_data = pd.read_csv(file_path, names=['metric', 'pearson', 'sample_size'])
        
        # Extract project ID and developer class from the filename
        match = pattern.match(filename)
        if match:
            project_id = match.group(1)
            developer_class = match.group(2)
            
            # Add columns for project ID and developer class
            project_data['project'] = project_id
            project_data['class'] = developer_class
            
            # Update the unique project count for each class
            if developer_class not in project_class_count:
                project_class_count[developer_class] = set()
            project_class_count[developer_class].add(project_id)
            
            # Append the data to the all_data DataFrame
            all_data = pd.concat([all_data, project_data], ignore_index=True)

# Convert the set of projects to their counts
for developer_class in project_class_count:
    project_class_count[developer_class] = len(project_class_count[developer_class])

# Transform the Pearson coefficient values into correlation levels
all_data['correlation_level'] = all_data['pearson'].apply(correlation_level)
all_data = all_data[all_data['metric'] != 'delta_action_delta_date']

# Count the number of times each correlation level appears for each metric and developer class
project_counts = all_data.groupby(['class', 'metric', 'correlation_level']).size().reset_index(name='num_projects')

# Merge the total number of projects into the project_counts dataframe
project_counts['total_projects'] = project_counts['class'].map(project_class_count)

# Calculate the percentage
project_counts['percentage'] = (project_counts['num_projects'] / project_counts['total_projects']) * 100

# Pivot the data to get the desired format
pivoted_data = project_counts.pivot_table(index='metric', columns=['correlation_level', 'class'], values='percentage', fill_value=0)
pivoted_data = pivoted_data.reindex(columns=pd.MultiIndex.from_product([['low', 'medium', 'strong'], ['super inactive', 'inactive', 'average', 'active', 'super active']]), fill_value=0)

# Split the data into separate tables for each correlation level
low_table = pivoted_data['low'].reset_index()
medium_table = pivoted_data['medium'].reset_index()
strong_table = pivoted_data['strong'].reset_index()

# Save the result to CSV files if needed
low_table.to_csv('correlation_levels_summary_low.csv', index=False)
medium_table.to_csv('correlation_levels_summary_medium.csv', index=False)
strong_table.to_csv('correlation_levels_summary_strong.csv', index=False)

# Save the result to LaTeX files if needed
low_table.to_latex('correlation_levels_summary_low.tex', index=False)
medium_table.to_latex('correlation_levels_summary_medium.tex', index=False)
strong_table.to_latex('correlation_levels_summary_strong.tex', index=False)

# Display the result
print("Low Correlation Table:")
print(low_table)
print("\nMedium Correlation Table:")
print(medium_table)
print("\nStrong Correlation Table:")
print(strong_table)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import re

directory_path = 'correlations'
pattern = re.compile(r'\d+_(\d+)_(.*?)\.csv')
project_class_count = {}
cmap = plt.get_cmap('Pastel1')
colors = [cmap(i) for i in range(0,2)][1]

for filename in os.listdir(directory_path):
    if filename.endswith(".csv"):  # Ensure the file is a CSV
        file_path = os.path.join(directory_path, filename)
        project_data = pd.read_csv(file_path, names=['metric', 'pearson', 'sample_size'])
        
        # Extract project ID and developer class from the filename
        match = pattern.match(filename)
        if match:
            project_id = match.group(1)
            developer_class = match.group(2)
            
            project_data['class'] = developer_class
            
            # Update the unique project count for each class
            if developer_class not in project_class_count:
                project_class_count[developer_class] = 1
            project_class_count[developer_class] = project_class_count[developer_class] + 1
            

df = pd.DataFrame(project_class_count.items(), columns=['class', 'count'])
df.plot.bar(x='class', y='count', title='Number of Projects per Class', xlabel='Class', ylabel='Number of Projects', rot=45, color=colors)
plt.style.use('default')


# plt.grid(None)
plt.savefig('number_of_projects_per_devs_class.png', bbox_inches='tight')
plt.savefig('number_of_projects_per_devs_class.pdf', bbox_inches='tight')

plt.show()
            

            
```
